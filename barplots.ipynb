```python
import pandas as pd
import matplotlib.pyplot as plt

data={"store_NO":["store_1","store_2","store_3","store_4"],
      "sales":[20,15,7,12]

}
df=pd.DataFrame(data)

plt.bar(df["store_NO"],df["sales"])
plt.xlabel('store number')
plt.ylabel("sales")
plt.title("bar plot of sales against store number")
plt.show()
```

```python
data={"store_NO":["store_1","store_2","store_3","store_4"],
      "Q1_sales":[20,15,7,12],
      "Q2_sales":[16,32,12,20]

}
df_1=pd.DataFrame(data)
df_1
```

```python
x=range(len(df_1["store_NO"]))
width=0.35

plt.bar(x,df_1["Q1_sales"],width,label="Q1_sales")
plt.bar([i+width for i in x],df_1["Q2_sales"],width,label="Q2_sales")

plt.xlabel("store_number")
plt.ylabel("sales")
plt.title("grouped bar plot of the sales")

plt.xticks([i+width/2 for i in x],df_1["store_NO"])

plt.legend()

plt.show()
```

```python
data={"store_NO":["store_1","store_2","store_3","store_4"],
      "Q1_sales":[20,15,7,12],
      "Q2_sales":[8,10,6,9]

}
df_1=pd.DataFrame(data)

plt.bar(df_1['store_NO'],df_1["Q1_sales"],label="Q1_sales")
plt.bar(df_1["store_NO"],df_1["Q2_sales"],bottom=df_1["Q2_sales"],label="Q2_sales")

plt.xlabel('store number')
plt.ylabel("sales")
plt.title("stacked bar plot of sales")


plt.legend()
plt.show()
```

```python
dataset=pd.DataFrame({
    "product":["electronics","veges","category"],
    "sales_overtime":[[10,12,14,16],[7,9,11,13,],[5,6,7,8]]
}

)
plt.violinplot(dataset["sales_overtime"],showmedians=True)
plt.xticks(range(1, len(dataset["product"])+1),dataset["product"])
plt.xlabel("sales")
plt.ylabel("product")
plt.title("violin plot")

plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_1 = pd.DataFrame({
    "category":["A","A","B","A","B","B"],
    "value":[10,15,7,12,9,8,]

}

)

sns.kdeplot(x=data_1["value"],hue=data_1["category"],fill=True)

plt.xlabel("value")
plt.ylabel("density")
plt.title("kernel density plot")
```
