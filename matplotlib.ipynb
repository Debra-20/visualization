```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [10, 15, 13, 18, 20]  # Y-axis values

# Create a line plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')

# Display the plot
plt.show()

```

```python

import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Scatter Plot')
plt.show()

```

```python

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [250, 350, 200, 450]

plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.show()

```

```python

exam_scores = [78, 85, 92, 67, 88, 75, 95, 84, 72, 90]

plt.hist(exam_scores, bins=5, edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.title('Exam Score Distribution')
plt.show()

#bins=5: The bins parameter specifies the number of bins or intervals into which the data will be divided in order to create the histogram. In your case, you've set it to 5, which means the exam scores will be grouped into 5 intervals.

#edgecolor='black': The edgecolor parameter determines the color of the edges or borders of the bars in the histogram. In your code, you've set it to 'black', which means the bars will have black borders.
```
