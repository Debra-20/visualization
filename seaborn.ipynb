```python
df=sns.load_dataset("titanic")
df.head(2)
```

```python
df.info()
```

```python
df.shape
```

```python
#bar plot
#syntax(sns.barplot(x="column"))
sns.barplot(x="sex",y="fare",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="fare",data=df,palette="Set2")
plt.show()
```

```python
#countplot

sns.countplot(x="sex",data=df,palette="Set1")
ax = sns.countplot(x="sex", data=df, palette="Set1")
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
#heatmap
sns.heatmap(df.corr(),cbar=True,linewidth=0.5)
plt.show()
```

```python
#scatter plot
sns.scatterplot(x="class",y="fare",data=df,palette="Set2")
plt.show()
```

```python
sns.pairplot(df)
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris,hue="species",palette="Set2")
plt.show
```

```python
#lmplot
sns.lmplot(x="petal_length",y="petal_width",data=iris)
plt.show()
```

```python
sns.lmplot(x="petal_length",y="petal_width",hue="species",data=iris,markers=["o","*","<"])
plt.show()
```

```python
#boxplot
sns.boxplot(x="species",y="sepal_width",data=iris)
plt.show()
```
